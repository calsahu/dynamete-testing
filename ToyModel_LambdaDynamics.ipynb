This notebook is designed to run lambda dynamics for the toy model where
$$ R = e^{-\lambda_1 n - \lambda_2 f(n)}/Z $$
and
$$ f(n) = b_0 n - d_0 N/N_c n^2. $$

The idea is to eventually test my derivation from information geometry against this lambda dynamics iteration. Note that the f function I've written I actually have changed a few times. This notebook has been run with N/N_c present and without it. Without it I think I can show lambda dynamics and the information geometry approach are the same, however with the N/N_c included I can't get it to work.

# Discussion
So in the end lambda dynamics is still approximating the brute force iteration, and the information geometry approach is not the same.

However, it's still unclear to me that we know that lambda dynamics is the correct iteration scheme over the information geometry approach at all. I was just really hoping they would be the same.
