This notebook is designed to run lambda dynamics for the toy model where
$$ R = e^{-\lambda_1 n - \lambda_2 f(n)}/Z $$
and
$$ f(n) = b_0 n - d_0 N/N_c n^2. $$

The idea is to eventually test my derivation from information geometry against this lambda dynamics iteration. Note that the f function I've written I actually have changed a few times. This notebook has been run with N/N_c present and without it. Without it I think I can show lambda dynamics and the information geometry approach are the same, however with the N/N_c included I can't get it to work.

```python
import numpy as np
import pandas as pd
import lambda_dynamics_ToyModel as ld
import DynaMETE_Rfunctions_ToyModel as rf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set up unperturbed parameters

# state variables
s = pd.Series(np.array([50, 10000]), index=['S','N']) #'S' = array.item(0), 'N' is array.item(1)

# parameters. We will derive the precise ones below
p_d = np.array([0.03,0.000075188,s['N']]) #s['N'] gives us array.item(1) (array is the first argument to pd.Series)
p = pd.Series(p_d,index=['b0','d0','Nc']) #to assign the birth rate, unperturbed death rate, and carrying capacity

# To get precise parameters
pp = ld.get_ss_params(s,p)

# Initial lambdas
li = rf.lambda_i(s)
# Display parameters and show derivatives are close enough to zero. (-900/2.5*10^5 is still small)
print("Non exact parameters")
display(p)
display(rf.get_dXdt(li,s,p))
print("Exact parameters")
display(pp)
display(rf.get_dXdt(li,s,pp))
```

### Perturbations
Let's try $d_0 \to 1.25 d_0$.
Let's try going out to about $t = 10$ or 20 with a relatively small (maybe 0.1 or 0.05) iteration step. I used 0.2 above just to see roughly what happens, but with the perturbation I think a smaller step size would be more appropriate.

I changed the perturbed $d_0$ to $0.00008$.

```python
perturb = pp.copy()
pdeathrate = (0.00008)/(0.000075188)
perturb['d0'] = perturb['d0']*pdeathrate #setting new rate such that the death rate is 0.00008 in the first iteration.
```

```python
l_d0,s_d0,ds_d0 = ld.iterate(200,s,perturb, dt=1)
```

```python
#Plotting the N-independent transition function
s_d0['N'].plot()
```

```python
ds_d0.plot()
```

```python
plt.plot(l_d0[:,0],label='L1')
plt.plot(l_d0[:,1],label='L2')
plt.legend()
#Closest to figure 2 of entropic dynamics paper
```

```python
# I expected l2 to be < 0 because of the n^2 term. What does R look like?
nr = np.arange(s['N'])+1
plt.plot(nr,rf.R(nr,l_d0[200],s_d0.iloc[200],perturb))
```

```python
#Setting up the simple logistic model - parameters

big_n = s['N']/s['S'] 
big_n_copy = s['N']/s['S']
f_n = 2*(big_n ** 2) - (big_n)
birth_rate = pp['b0']
array_of_ns = [50 * big_n]
perturbation = 0.00008

#iterations

for i in np.arange(200):
    f_n = 2 * (big_n ** 2) - (big_n)
    growth = big_n * birth_rate
    decline = perturbation * f_n * big_n/big_n_copy
    big_n = big_n + growth - decline
    true_big_n = 50 * big_n
    array_of_ns = np.append(array_of_ns, true_big_n)
    perturbation = perturbation * pdeathrate
    
plt.plot(array_of_ns)
```

```python
#plotting the N-independent transition function(without N/Nc term)
```

```python
#plotting both the logistic function and the N-independent transition functions together:
plt.plot(array_of_ns,label='Logistic Function')
plt.plot(s_d0['N'],label='N-dependent transition function')
plt.legend()
plt.ylim([9000, 10000])

#Compared with Figure 1 of Entropic Dynamics and Updating of Lagrange Multipliers Paper
```

# Discussion
So in the end lambda dynamics is still approximating the brute force iteration, and the information geometry approach is not the same.

However, it's still unclear to me that we know that lambda dynamics is the correct iteration scheme over the information geometry approach at all. I was just really hoping they would be the same.

```python
# Look at all three dN plots
plt.plot(ds_d0['dN'])
plt.plot(ds_d0g['dN'])
plt.plot(ds_d0bf['dN'])
```

```python
# Let's try to look at d^2N/dt^2, since that may be where the difference is coming.
plt.plot(np.gradient(ds_d0bf['dN']))
```

```python
# Now calculate cov(f,f) dl2/dt + cov(n,f) dl1/dt at each time step
dn2t = []#np.zeros([len(ds_d0bf['dN'])])
dl1t = np.gradient(l_d0bf[:,0])
dl2t = np.gradient(l_d0bf[:,1])
for lt,st,dst,dl1tt,dl2tt in zip(l_d0bf,s_d0bf.iterrows(),ds_d0bf.iterrows(),dl1t,dl2t):
    zt = rf.mean_pow(0,lt,st[1],perturb)
    covff = rf.cov(rf.f,rf.f,lt,st[1],perturb,zt)
    covnf = rf.cov(lambda n,s,p: n,rf.f,lt,st[1],perturb,zt)
    mdfdt = rf.mean(rf.dfdt,lt,st[1],perturb,dst[1],z=zt)
    covfdf = rf.cov(lambda n,s,p,ds: rf.f(n,s,p),rf.dfdt,lt,st[1],perturb,zt,dst[1])
    #dn2t.append(-covnf*dl1tt-covff*dl2tt - covfdf*lt[1] + mdfdt)
    dn2t.append(-covnf*dl1tt-covff*dl2tt)# - covfdf*lt[1] + mdfdt)
```
